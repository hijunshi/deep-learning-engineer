##  Linear Regression & Loss Function

```python
%matplotlib inline
import numpy as np

import matplotlib.pyplot as plt
```

## 문제 1 Grid Search를 활용한 Linear Regression

linear regression 과 **Grid Search**  을 이용해 아래 데이터를 linear regression 해주세요. 


###  Grid Search


1. 탐색할 가중치 조합을 구성합니다. 이 때 `np.meshgrid` 혹은 `np.mgrid`로 구성합니다. 

2. 위의 가중치 조합에서 가중치를 뽑아, 예측값을 산출합니다.

3. 실제 값과 예측값의 차이를 MSE로 계산합니다.

3. 모든 가중치 조합 중 가장 작은 MSE를 보이는 Loss값 조합을 찾습니다.


```python
# 아래 코드를 실행해 주세요 
num_points = 1000
vectors_set = []
for i in range(num_points):
    x1= np.random.normal(0.0, 0.55)
    y1= x1 * 0.1 + 0.3 + np.random.normal(0.0, 0.03)
    vectors_set.append([x1, y1])

xs = np.asarray([v[0] for v in vectors_set])
ys = np.asarray([v[1] for v in vectors_set])

plt.plot(xs, ys, 'ro', label='Original data')
plt.legend()
plt.show()
```

#### 답안지 입력란
* 아래에 답을 서술하여 주세요

```python
# 정답을 서술해주세요!

# Grid Search 좌표준비
ws = np.linspace(0,2,1000)
bs = np.linspace(0,1,1000)
# 좌표 Grid
mg_ = np.meshgrid(ws, bs)
# 좌표점 array 변환
prms = (np.stack(mg_, axis=-1)).reshape(-1,2)
# prms.shape ~ (1000000, 2)

# cost function 정의
def get_loss(w, b):
    yhat = w * xs + b
    return ((ys - yhat)**2).mean()

# losses 값 산출
losses = []
for w, b in prms:
    losses.append(get_loss(w,b))
losses = np.array(losses)


# losses가 최소인 인덱스 정의
sorted_idx = losses.argsort()
# loss 최소점인 좌표점 추출
sorted_labels = prms[sorted_idx]

# what, bhat=sorted_labels[:1]
# slop
what = sorted_labels[:1,0]
# 절편
bhat = sorted_labels[:1,1]


# 결과 그리기
plt.scatter(xs, ys, c='y', marker='o', label='true')
plt.plot(xs, what*xs+bhat, label='pred')
plt.legend()
plt.show()
```

## 문제 2

문제 1과 동일합니다. 다만 Loss function은 MAE(mean absolute Error)을 사용합니다 <br>
아래와 같은 그림이 나올수 있도록 Linear Regression 결과를 시각화 해주세요 

```python
# 아래 코드를 실행해 주세요 
num_points = 1000
vectors_set = []
for i in range(num_points):
    x1= np.random.normal(0.0, 0.55)
    y1= x1 * 0.1 + 0.3 + np.random.normal(0.0, 0.03)
    vectors_set.append([x1, y1])

xs = np.asarray([v[0] for v in vectors_set])
ys = np.asarray([v[1] for v in vectors_set])

plt.plot(xs, ys, 'ro', label='Original data')
plt.legend()
plt.show()
```

#### 답안지 입력란
* 아래에 답을 서술하여 주세요

```python
# 정답을 서술해 주세요!
# Grid Search 좌표준비
ws = np.linspace(0,2,1000)
bs = np.linspace(0,1,1000)
# 좌표 Grid
mg_ = np.meshgrid(ws, bs)
# 좌표점 array 변환
prms = (np.stack(mg_, axis=-1)).reshape(-1,2)
# prms.shape ~ (1000000, 2)


# cost function 정의
def get_loss(w, b):
    yhat = w * xs + b
    return (np.absolute(ys-yhat)).mean()

# losses 값 산출
losses = []
for w, b in prms:
    losses.append(get_loss(w,b))
losses = np.array(losses)


# losses가 최소인 인덱스 정의
sorted_idx = losses.argsort()
# loss 최소점인 좌표점 추출
sorted_labels = prms[sorted_idx]

# what, bhat=sorted_labels[:1]
# slop
what = sorted_labels[:1,0]
# 절편
bhat = sorted_labels[:1,1]


# 결과 그리기
plt.scatter(xs, ys, c='y', marker='o', label='true')
plt.scatter(xs, what*xs+bhat, marker='o', label='pred')
plt.legend()
plt.show()
```
